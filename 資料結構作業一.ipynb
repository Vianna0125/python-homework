```python
user = {}
while True:
  choise = input('請選擇操作選項(a註冊，b登入，c退出)')
  if choise == 'a':
    username = input('請輸入帳號')
    if username in user:
      print('帳號已存在')
    else:
      password = input('請輸入密碼')
      user[username] = password
      print('註冊成功!')
  elif choise == 'b':
    username = input('請輸入帳號')
    password = input('請輸入密碼')
    if username in user and user[username] == password:
      print('登入成功!')
    else:
      print('登入失敗')
  elif choise == 'c':
    print('退出程式')
    break
  else:
    print('無效的操作，重新輸入')
```
